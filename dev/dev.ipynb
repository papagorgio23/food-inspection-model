```python
# group by Restaurant_Name and calculate the number of days between inspections
df['Inspection_Date'] = pd.to_datetime(df['Inspection_Date'])


# Sort the DataFrame by store name and inspection date
df = df.sort_values(by=["Restaurant_Name", "Inspection_Date"])

# Calculate the number of days since the previous inspection_date
df["days_since_previous_inspection"] = df.groupby("Restaurant_Name")["Inspection_Date"].diff().dt.days

# Replace missing values in the new column with 0 (for the first inspection)
df["days_since_previous_inspection"] = df["days_since_previous_inspection"].fillna(0).astype(int)

df
```

```python
# Sort the DataFrame by store name and inspection date
df = df.sort_values(by=["Restaurant_Name", "Inspection_Date"])

# Calculate the previous number of violations (Violations)
df["previous_demerits"] = df.groupby("Restaurant_Name")["Inspection_Demerits"].shift(fill_value=0)

df
```

```python
# Sort the DataFrame by store name and inspection date
df = df.sort_values(by=["Restaurant_Name", "Inspection_Date"])

# Create a binary flag column indicating if it's the first inspection for each store
df["is_first_inspection"] = (df.groupby("Restaurant_Name").cumcount() == 0).astype(int)
df
```

# Model Work

```python
# Target variable is 1 if the inspection grade is 'A', else 0
target_variable = 'target_variable'

# Features to be used in the model
predictors = ['days_since_previous_inspection', 'Inspection_Type', 'is_first_inspection', 'previous_demerits', 'Category_Name', 'Current_Grade', 'Current_Demerits']
```

```python
model_data = df[[target_variable] + predictors]
model_data.info()
```

```python
model_data.head()
```

If I had more time i would look into the missing values closer but to get this done quickly I will just drop them.

```python
# drop rows with missing values
model_data = model_data.dropna()
model_data.info()
```

```python
# convert categorical variables to dummies using one-hot encoding (i'm keeping all the columns because I'm going to use a tree-based model)
model_data = pd.get_dummies(model_data, columns=['Inspection_Type', 'Category_Name', 'Current_Grade'], drop_first=False, dtype=int)
model_data.head()
```

```python
model_data.describe()
```

```python
# Compute the correlation matrix
corr = model_data.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# reset index
model_data = model_data.reset_index(drop=True)
```

```python
model_data.info()
```

```python
# create xgboost model
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


X = model_data.drop(target_variable, axis=1)
y = model_data[target_variable]

# split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# split training data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.3, random_state=42)


model = xgb.XGBClassifier(objective='binary:logistic', n_estimators=1000, seed=123)
model.fit(X_train, y_train, early_stopping_rounds=5, eval_set=[(X_val, y_val)], verbose=True)

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f'Accuracy: {accuracy}')
```

```python
# feature importance
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 8))
xgb.plot_importance(model, ax=ax)
plt.show()
```

```python
# save model
import joblib

joblib.dump(model, '../models/model.pkl')
```
